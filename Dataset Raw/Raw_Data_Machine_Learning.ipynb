```python
# -----------------------------
# 1. Mount Google Drive
# -----------------------------
from google.colab import drive
drive.mount('/content/drive')

# Folder tujuan (ganti jika mau)
save_path = "/content/drive/MyDrive/IMDB_Raw_Dataset/"
import os
os.makedirs(save_path, exist_ok=True)

# -----------------------------
# 2. Load IMDB Dataset (HuggingFace)
# -----------------------------
from datasets import load_dataset
import pandas as pd

dataset = load_dataset("imdb")

train_texts = dataset["train"]["text"]
train_labels = dataset["train"]["label"]

test_texts = dataset["test"]["text"]
test_labels = dataset["test"]["label"]

# -----------------------------
# 3. Convert to DataFrame
# -----------------------------
df_train = pd.DataFrame({
    "text": train_texts,
    "label": train_labels
})

df_test = pd.DataFrame({
    "text": test_texts,
    "label": test_labels
})

# -----------------------------
# 4. Save to Google Drive as CSV
# -----------------------------
train_file = save_path + "imdb_raw_train.csv"
test_file = save_path + "imdb_raw_test.csv"

df_train.to_csv(train_file, index=False, encoding='utf-8')
df_test.to_csv(test_file, index=False, encoding='utf-8')

print("Dataset berhasil disimpan ke Google Drive!")
print("Train CSV:", train_file)
print("Test CSV:", test_file)
```
